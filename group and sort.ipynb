```python
minprice = reviews.groupby("variety").price.min()
maxprice = reviews.groupby("variety").price.max()
price_extremes = minprice.add(maxprice)
price_extremes = price_extremes.sort_index()
price_extremes
```

```python
price_extremes = reviews.groupby('variety').price.agg([min, max])
price_extremes
```

```python
sorted_varieties = price_extremes.sort_values(by=["min", "max"], ascending=False)
sorted_varieties
```
